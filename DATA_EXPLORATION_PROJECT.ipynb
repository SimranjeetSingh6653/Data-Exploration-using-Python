# Reading and Previewing Diabetes Dataset
Reads "Diabetes.csv" into a DataFrame and displays the initial rows for data inspection.

```python
df = pd.read_csv("Diabetes.csv")
df.head()
```

# Violin Plot with Swarm Plot Overlay
Displays the distribution of BMI by class using a violin plot, with individual data points overlaid as a swarm plot, aiding in understanding class-wise variation in BMI.

```python
# Violin plot with swarm plot overlay
plt.figure(figsize=(10, 8))
sns.violinplot(x='class', y='BMI', data=df, inner=None)
sns.swarmplot(x='class', y='BMI', data=df, color='k', alpha=0.5)
plt.title('BMI Distribution by Class')
plt.xlabel('Class')
plt.ylabel('BMI')
plt.show()
```

# Joint Plot with Regression Lines

Insights: Visualizes the relationship between Plasma Glucose Concentration and Serum Insulin levels using a joint plot with regression lines, providing insights into their correlation.

```python
# Joint plot with regression lines
sns.jointplot(x='PG Concentration', y='Serum Ins', data=df, kind='reg', scatter_kws={'alpha':0.5})
plt.suptitle('PG Concentration vs. Serum Ins with Regression Lines')
plt.show()
```

# PairGrid with Kernel Density Estimates

Insights: Utilizes PairGrid to visualize pairwise relationships with kernel density estimates, offering insights into the distribution of variables and their interactions.

```python
# PairGrid with kernel density estimates
g = sns.PairGrid(df, hue='class')
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
g.map_diag(sns.kdeplot, lw=2)
g.add_legend()
plt.suptitle("PairGrid with Kernel Density Estimates")
plt.show()
```

# Ridgeline Plot of Age by Class

Insights: Utilizes a ridgeline plot to display the distribution of Age by class, with filled density curves, aiding in visualizing age distributions across different classes.

```python
# Ridgeline plot
plt.figure(figsize=(10, 8))
sns.kdeplot(data=df, x='Age', hue='class', fill=True, common_norm=False, palette='husl', alpha=0.5)
plt.title('Ridgeline Plot of Age by Class')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

# Correlation Matrix with Hierarchical Clustering

Insights: Displays a clustered heatmap of the correlation matrix with hierarchical clustering, revealing patterns of variable relationships through grouping, aiding in identifying correlated features.

```python
# Correlation matrix with hierarchical clustering
plt.figure(figsize=(10, 8))
sns.clustermap(df.corr(), cmap='coolwarm', figsize=(10, 8))
plt.title('Correlation Matrix with Hierarchical Clustering')
plt.show()
```

# Clustered Boxplot with Swarm Plot

Insights: Combines a clustered boxplot with a swarm plot to display the distribution of each feature, aiding in identifying outliers and understanding data spread across variables.

```python
# Clustered boxplot with swarm plot
plt.figure(figsize=(10, 8))
sns.boxplot(data=df, orient='h')
sns.swarmplot(data=df, orient='h', color='.25')
plt.title('Clustered Boxplot with Swarm Plot')
plt.show()
```

# Pairwise Relationships with Hexbin Plots

Insights: Utilizes hexbin plots in a pairplot to visualize the joint distribution of variables, offering a clearer view of data density and relationships between features.

```python
# Pairwise relationships with hexbin plots
sns.pairplot(df, hue='class', kind='hex', palette='husl')
plt.suptitle("Pairplot with Hexbin Plots")
plt.show()
```

```python
# Violin swarm plot matrix
plt.figure(figsize=(12, 10))
sns.violinplot(data=df, inner=None)
sns.swarmplot(data=df, color='k', alpha=0.5)
plt.title('Violin Swarm Plot Matrix')
plt.xticks(rotation=45)
plt.show()
```

# Violin Swarm Plot Matrix

Insights: Displays the distribution of each variable using violin plots, with individual data points overlaid as a swarm plot, providing a comprehensive overview of data distribution across multiple features.

```python
# Hexbin map
plt.figure(figsize=(10, 8))
plt.hexbin(df['Diastolic BP'], df['PG Concentration'], gridsize=20, cmap='coolwarm')
plt.colorbar(label='count in bin')
plt.xlabel('Diastolic BP')
plt.ylabel('PG Concentration')
plt.title('Hexbin Map of Diastolic BP vs. PG Concentration')
plt.show()
```
